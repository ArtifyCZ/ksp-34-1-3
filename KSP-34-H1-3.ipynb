```python
class Task:
    def __init__(self, orderId, points, deadlineDay):
        self._orderId = orderId
        self._points = points
        self._deadlineDay = deadlineDay
        self._doDay = -1
        pass
    
    def getOrderId(self):
        return self._orderId
    
    def getPoints(self):
        return self._points
    
    def getDeadlineDay(self):
        return self._deadlineDay
    
    def getDoDay(self):
        return self._doDay
    
    def setDoDay(self, doDay):
        self._doDay = doDay
        pass
    pass
```

```python
n = int(input())

tasks_in = [None] * n

for orderId in range(n):
    points, deadlineDay = input().split()
    points = int(points)
    deadlineDay = int(deadlineDay)

    task = Task(orderId, points, deadlineDay)
    tasks_in[orderId] = task
    pass
```

```python
def sortUngroupedTasksByDeadline(tasks):
    return sorted(tasks, key=lambda t: t.getDeadlineDay(), reverse=True)

def sortUngroupedTasksByPoints(tasks):
    return sorted(tasks, key=lambda t: t.getPoints(), reverse=True)

def sortUngroupedTasksByOrderId(tasks):
    return sorted(tasks, key=lambda t: t.getOrderId())

tasks_in = sortUngroupedTasksByDeadline(tasks_in)
```

```python
def groupTasks(tasks):
    res = [[]]
    
    i = 0
    j = 0
    dd = tasks[0].getDeadlineDay()
    
    for t in tasks:
        if t.getDeadlineDay() < dd:
            res[i] = sorted(res[i], key=lambda t2: t2.getPoints(), reverse=True)
            res.append([])
            i += 1
            j = 0
            dd = t.getDeadlineDay()
            pass
        res[i].append(t)
        j += 1
        pass
    
    return res

tasks = groupTasks(tasks_in)
```

```python
def getNextDeadline(tasks, i):
    if i == (len(tasks) - 1):
        return -1
    return tasks[i + 1][0].getDeadlineDay()

def setDoDays(tasks):
    i = 0
    j = 0
    a = tasks[0]
    b = []
    c = []
    
    for t1 in tasks:
        for t2 in t1:
            c.append(t2)
            pass
        pass
    
    k = getNextDeadline(tasks, -1)
    nextDeadline = getNextDeadline(tasks, 0)
    while True:
        if k == -1:
            break
            pass
        if len(a) == 0:
            k = nextDeadline
            pass
        if k == nextDeadline:
            j += 1
            a += tasks[j]
            a = sortUngroupedTasksByPoints(a)
            nextDeadline = getNextDeadline(tasks, j)
            pass
        k -= 1
        t = a[0]
        a.pop(0)
        b.append(t)
        pass
    
    b.reverse()
    j = 0
    
    for t in b:
        t.setDoDay(j)
        j += 1
        pass
    
    return c
```

```python
tasks = setDoDays(tasks)
total = 0
for t in tasks:
    if t.getDoDay() != -1:
        total += t.getPoints()
    pass

print(total)
tasks = sortUngroupedTasksByOrderId(tasks)

for t in tasks:
    if t.getDoDay() > t.getDeadlineDay():
        raise ValueError("DO DAY {} WAS GREATER THAN DEADLINE {}".format(t.getDoDay(), t.getDeadlineDay()))
        pass
    print(t.getDoDay())
    pass

# for t in tasks:
#     for t2 in t:
#         print("order id: {}; points: {}; day: {}; deadline: {}".format(t2.getOrderId(), t2.getPoints(), t2.getDoDay(), t2.getDeadlineDay()))
#         pass
#     pass
```
